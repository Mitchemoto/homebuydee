```python
import requests
import pandas as pd
from datetime import datetime

# Read the API key from the text file
api_key_file = r"C:\Users\mitchele\University of Iowa\Dedas, Jason M - Analytics Experience Group 2\Code\rapidapi_key.txt"
with open(api_key_file, 'r') as file:
    rapidapi_key = file.read().strip()

# Define the URLs and headers
url = "https://zillow-com1.p.rapidapi.com/propertyExtendedSearch"
url1 = "https://zillow-com1.p.rapidapi.com/property"
host = 'zillow-com1.p.rapidapi.com'

# Define the first query string to get ZPIDs
queryString0 = {
	"location" = "ct",
	"page" = "20",
	"status_type" = "RecentlySold",
	"home_type" = "Houses, Condos, Townhomes",
	"minPrice" = "100000",
	"maxPrice" = "119999.99",
	"soldInLast" = "36m",
    "isForSaleForeclosure": "false"
}

queryString1 = {
	"location" = "ct",
	"page" = "20",
	"status_type" = "RecentlySold",
	"home_type" = "Houses, Condos, Townhomes",
	"minPrice" = "120000",
	"maxPrice" = "139999.99",
	"soldInLast" = "36m",
    "isForSaleForeclosure": "false"
}

queryString2 = {
	"location" = "ct",
	"page" = "20",
	"status_type" = "RecentlySold",
	"home_type" = "Houses, Condos, Townhomes",
	"minPrice" = "140000",
	"maxPrice" = "159999.99",
	"soldInLast" = "36m",
    "isForSaleForeclosure": "false"
}

queryString3 = {
	"location" = "ct",
	"page" = "20",
	"status_type" = "RecentlySold",
	"home_type" = "Houses, Condos, Townhomes",
	"minPrice" = "160000",
	"maxPrice" = "179999.99",
	"soldInLast" = "36m",
    "isForSaleForeclosure": "false"
}

queryString4 = {
	"location" = "ct",
	"page" = "20",
	"status_type" = "RecentlySold",
	"home_type" = "Houses, Condos, Townhomes",
	"minPrice" = "180000",
	"maxPrice" = "199999.99",
	"soldInLast" = "36m",
    "isForSaleForeclosure": "false"
}


headers = {
    'x-rapidapi-key': rapidapi_key,
    'x-rapidapi-host': host
}

# Make the first API call to get ZPIDs
response = requests.get(url, headers=headers, params=queryString)

# Check the response status
if response.status_code == 200:
    # Parse the response content to extract properties
    content_response = response.json()
    
    # Debug: Print the response to understand its structure
    print("Content Response:", content_response)
    
    # Extract the list of properties
    properties = content_response.get('props', [])
    
    # Initialize lists to store property details
    zpid_list = []
    date_sold_list = []
    property_details_list = []
    
    # Loop through each property to extract details
    for prop in properties:
        zpid = prop.get('zpid')
        date_sold = datetime.utcfromtimestamp(prop.get('dateSold') / 1000).strftime('%Y-%m-%d') if 'dateSold' in prop else None
        
        # Collect basic property info
        zpid_list.append(zpid)
        date_sold_list.append(date_sold)
        property_details_list.append(prop)
    
    # Combine ZPIDs and dateSold into a DataFrame
    property_info = pd.DataFrame({
        'zpid': zpid_list,
        'dateSold': date_sold_list
    })
    
    # Debug: Print the property_info DataFrame to verify content
    print("Property Info DataFrame:", property_info)
    
    # Initialize a list to store detailed property information
    detailed_property_info = []

    # Loop through each ZPID and make the second API call to get detailed property information
    for zpid in zpid_list:
        queryString1 = {"zpid": zpid}
        
        response1 = requests.get(url1, headers=headers, params=queryString1)
        
        # Check the response status
        if response1.status_code == 200:
            # Parse the response content
            property_detail = response1.json()
            # Add zpid and dateSold to the property_detail
            property_detail['zpid'] = zpid
            property_detail['dateSold'] = property_info[property_info['zpid'] == zpid]['dateSold'].values[0]
            # Append the detailed property information to the list
            detailed_property_info.append(property_detail)
        else:
            print(f"Failed to get details for ZPID: {zpid}")
    
    # Debug: Print detailed_property_info to verify content
    print("Detailed Property Info List:", detailed_property_info)
    
    # Convert the list of detailed property information to a DataFrame
    if detailed_property_info:
        property_df = pd.json_normalize(detailed_property_info)
    else:
        property_df = pd.DataFrame()
    
    # Print the detailed property information as a DataFrame
    print("Property Details DataFrame:", property_df)
else:
    print("Failed to get ZPIDs")
```
