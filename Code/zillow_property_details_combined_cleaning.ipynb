```python
# parse out specific information for each zpid from zillow_by_zpid_property_details.csv and create zillow_by_zpid_property_details.csv
import pandas as pd

# Define the path to the CSV file
csv_path = r"...App/Data/zillow_property_details_combined.csv"

# Read the CSV file into a DataFrame
properties_df = pd.read_csv(csv_path)

# Select columns for review
prop_df = properties_df[['zpid',
                         'homeStatus',
                         'homeType',
                         'propertyTypeDimension',
                         'streetAddress',
                         'city',
                         'state',
                         'zipcode',
                         'county',
                         'propertyTaxRate',
                         'dateSold',
                         'timeOnZillow',
                         'price',
                         'zestimate',
                         'zestimateLowPercent',
                         'zestimateHighPercent',
                         'url',
                         'imgSrc',
                         'description',
                         'yearBuilt',
                         'livingArea',
                         'livingAreaValue',
                         'livingAreaUnits',
                         'bedrooms',
                         'bathrooms',
                         'annualHomeownersInsurance',
                         'monthlyHoaFee',
                         'longitude',
                         'latitude',
                         'favoriteCount',
                         'pageViewCount'
                         ]]

# Function to convert timeOnZillow to days
def convert_to_days(time_str):
    if 'days' in time_str or 'day' in time_str:
        return float(time_str.replace(' days', '').replace(' day', ''))
    elif 'hours' in time_str or 'hour' in time_str:
        return float(time_str.replace(' hours', '').replace(' hour', '')) / 24
    else:
        return float(time_str)  # Default case, just in case

# Rename and clean the timeOnZillow column
prop_df = prop_df.rename(columns={'timeOnZillow': 'timeOnZillowDays'})
prop_df['timeOnZillowDays'] = prop_df['timeOnZillowDays'].apply(convert_to_days)

# Drop rows where 'homeType' is 'HOME_TYPE_UNKNOWN' or 'LOT'
prop_df = prop_df[~prop_df['homeType'].isin(['HOME_TYPE_UNKNOWN', 'LOT'])]

# Drop rows where 'homeType' is 'HOME_TYPE_UNKNOWN' or 'LOT'
prop_df = prop_df[~prop_df['homeStatus'].isin(['FOR_RENT'])]

# Define the output path for the new CSV file
output_csv_path = r"...App/Data/zillow_by_zpid_property_details.csv"

# Save the selected columns to a new CSV file
prop_df.to_csv(output_csv_path, index=False)

print("Data has been processed and saved to", output_csv_path)
```
