```python
import pandas as pd

# Define the paths to the CSV files
# Full file path hidden
csv_path = r"...App/Data/property_details.csv"
csv_path1 = r"...App/Data/property_details_missing.csv"

# Read the CSV files into DataFrames
properties_df = pd.read_csv(csv_path)
details_df = pd.read_csv(csv_path1)

# Drop columns that contain only null values in each DataFrame
properties_df = properties_df.dropna(axis=1, how='all')
details_df = details_df.dropna(axis=1, how='all')

# Combine the two DataFrames
combined_df = pd.concat([properties_df, details_df], ignore_index=True)

# Drop rows that contain only null values
combined_df = combined_df.dropna(axis=0, how='all')

# Define the path to save the combined CSV
csv_path_combined = r"...App/Data/property_details_combined.csv"

# Save the combined DataFrame to a CSV file
combined_df.to_csv(csv_path_combined, index=False)

print(f"Combined property details saved to {csv_path_combined}")

```
